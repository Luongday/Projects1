```python
import pandas as pd
from datasets import Dataset, concatenate_datasets
#Load dataset
files = [
    "20Sentence1.csv",
    "20Sentence2.csv",
    "20Sentence3.csv",
    "20Sentence4.csv",
    "20Sentence5.csv",
    "Experiment.csv",
    "15_LTMH.csv",
    "30DoanData.csv",
    "BoDuLieuTest.csv",
    "vn_dataset.csv",
    "Data1_vb_final.csv",
    "DatasetDV.csv",
    "STR.csv",
    "VieParagraph2.csv",
    "data.csv",
]

dfs = []
for file in files:
    df = pd.read_csv(file)
    # Chuẩn hóa
    df['Sentence1'] = df['Sentence1'].str.replace('_', ' ')
    df['Sentence2'] = df['Sentence2'].str.replace('_', ' ')
    df['Score'] = df['Level'] / 2.0 # Đưa về mức điểm 0 -> 1
    dfs.append(df[['Sentence1', 'Sentence2', 'Score']])
full_df = pd.concat(dfs, ignore_index=True).drop_duplicates()
# Loại bỏ các hàng có giá trị null để tránh lỗi float(None)
full_df = full_df.dropna(subset=['Sentence1', 'Sentence2', 'Score'])
print(f"Tổng mẫu sau khi gộp và làm sạch: {len(full_df)}")

# Chia tap train/valid/test
dataset = Dataset.from_pandas(full_df)
dataset = dataset.train_test_split(test_size=0.2, seed=42)
train_val = dataset['train']
test_ds = dataset['test']
train_val = train_val.train_test_split(test_size=0.125, seed=42)
train_ds = train_val['train']
val_ds = train_val['test']
print(f"Tổng mẫu: {len(full_df)}, Train: {len(train_ds)}, Val: {len(val_ds)}, Test: {len(test_ds)}")
```

```python
#fine-tune with sentence-transformers
from sentence_transformers import SentenceTransformer, losses, util, InputExample, models
from sentence_transformers import SentenceTransformerTrainer
from sentence_transformers.training_args import SentenceTransformerTrainingArguments
from datasets import Dataset
from transformers import AutoTokenizer, AutoModel
import pandas as pd

# Load model
model = SentenceTransformer("dangvantuan/vietnamese-embedding")
model.max_seq_length = 256

def make_hf_dataset(ds):
    data = {
        "sentence1": [str(row['Sentence1']) for row in ds],
        "sentence2": [str(row['Sentence2']) for row in ds],
        "score": [float(row['Score']) for row in ds],
    }
    return Dataset.from_dict(data)

train_hf = make_hf_dataset(train_ds)
val_hf = make_hf_dataset(val_ds)

train_loss = losses.CosineSimilarityLoss(model=model)

args = SentenceTransformerTrainingArguments(
    output_dir = "vietnamese-embedding-relatedness-600",
    num_train_epochs=2,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=10,
    fp16=False,
    use_cpu=True,
    eval_strategy="no",
    save_strategy="epoch",
    logging_steps=100,
    save_total_limit=2,
    load_best_model_at_end=False,
    report_to="none",
)

trainer = SentenceTransformerTrainer(
    model=model,
    args=args,
    train_dataset=train_hf,
    eval_dataset=val_hf,
    loss=train_loss,
)

trainer.train()
model.save("vietnamese-embedding-relatedness-v1")
```

```python
from scipy.stats import spearmanr
import torch

preds, labels = [], []
for row in val_hf:
  emb1 = model.encode(row['sentence1'], convert_to_tensor=True)
  emb2 = model.encode(row['sentence2'], convert_to_tensor=True)
  cos_sim = torch.cosine_similarity(
      emb1.unsqueeze(0),
      emb2.unsqueeze(0)
  ).item()
  preds.append(cos_sim)
  labels.append(row['score'])
corr, _ = spearmanr(preds, labels)
print(f"Separman: {corr:.4f}")
```

```python
from sentence_transformers import SentenceTransformer, util
from scipy.stats import spearmanr
import numpy as np
import pandas as pd
from datetime import datetime

# 1. Load model đã train
model_path = "vietnamese-embedding-relatedness-v1"
model = SentenceTransformer(model_path)
print("Model loaded for testing!")

# 2. Chuẩn bị tập test
test_df = test_ds.to_pandas()

# Clean text nếu cần (đảm bảo không còn '_')
test_df['Sentence1'] = test_df['Sentence1'].astype(str).str.replace('_', ' ').str.strip()
test_df['Sentence2'] = test_df['Sentence2'].astype(str).str.replace('_', ' ').str.strip()

# 3. Tính predictions và lưu chi tiết từng cặp
results = []
preds = []
labels = []

for idx, row in test_df.iterrows():
    s1 = row['Sentence1']
    s2 = row['Sentence2']
    true_score = float(row['Score'])  # hoặc 'score' tùy tên cột

    # Encode và tính cosine similarity
    emb1 = model.encode(s1)
    emb2 = model.encode(s2)
    cos_sim = util.cos_sim(emb1, emb2)[0][0].item()

    preds.append(cos_sim)
    labels.append(true_score)

    # Lưu chi tiết
    results.append({
        'index': idx,
        'sentence1': s1,
        'sentence2': s2,
        'true_score': true_score,
        'predicted_similarity': round(cos_sim, 4),
        'difference': round(abs(true_score - cos_sim), 4)
    })

# 4. Tính metric tổng
corr, p_value = spearmanr(preds, labels)
mse = np.mean((np.array(preds) - np.array(labels)) ** 2)
mae = np.mean(np.abs(np.array(preds) - np.array(labels)))
binary_preds = [1 if p > 0.5 else 0 for p in preds]
binary_labels = [1 if l > 0.5 else 0 for l in labels]
acc = np.mean(np.array(binary_preds) == np.array(binary_labels))

# 5. Tạo DataFrame kết quả
results_df = pd.DataFrame(results)

# Summary metric
summary = {
    'Thời gian test': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    'Model': model_path,
    'Số cặp test': len(test_df),
    'Spearman correlation': f"{corr:.4f} (p-value: {p_value:.4f})",
    'MSE': f"{mse:.4f}",
    'MAE': f"{mae:.4f}",
    'Binary accuracy (threshold 0.5)': f"{acc:.4f}"
}

summary_df = pd.DataFrame([summary])

# 6. Lưu ra file Excel (2 sheet: Summary + Detailed)
output_file = "test_results_relatedness.xlsx"
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    summary_df.to_excel(writer, sheet_name='Summary', index=False)
    results_df.to_excel(writer, sheet_name='Detailed Results', index=False)

print(f"\nKẾT QUẢ ĐÃ ĐƯỢC LƯU VÀO FILE: {output_file}")
print("Mở file để xem Summary và chi tiết từng cặp (có thể sort theo difference để xem cặp sai nhiều nhất).")

# In summary ra màn hình để xem nhanh
print("\n=== SUMMARY ===")
for key, value in summary.items():
    print(f"{key}: {value}")

# In 5 ví dụ đầu để kiểm tra bằng mắt
print("\n=== 5 ví dụ đầu tiên ===")
for i in range(min(5, len(results_df))):
    row = results_df.iloc[i]
    print(f"Cặp {i+1}:")
    print(f"  Sentence1: {row['sentence1'][:80]}...")
    print(f"  Sentence2: {row['sentence2'][:80]}...")
    print(f"  True: {row['true_score']:.3f} | Predicted: {row['predicted_similarity']:.3f} | Diff: {row['difference']:.3f}")
    print("-" * 60)
```

```python
from sentence_transformers import SentenceTransformer, util
from scipy.stats import spearmanr
import numpy as np
import pandas as pd
import random
from datetime import datetime

# 1. Load model
model_path = "vietnamese-embedding-relatedness-v1"
model = SentenceTransformer(model_path)
print("Model loaded!")

# 2. Chuẩn bị test set
test_df = test_ds.to_pandas()
test_df = test_df.dropna(subset=['Sentence1', 'Sentence2', 'Score'])

# Clean text nếu cần
test_df['Sentence1'] = test_df['Sentence1'].astype(str).str.replace('_', ' ').str.strip()
test_df['Sentence2'] = test_df['Sentence2'].astype(str).str.replace('_', ' ').str.strip()

# 3. Chia ngẫu nhiên thành 2 nửa (split-half)
random.seed(42)  # để reproducible
test_shuffled = test_df.sample(frac=1).reset_index(drop=True)
half_size = len(test_shuffled) // 2

half1 = test_shuffled.iloc[:half_size]
half2 = test_shuffled.iloc[half_size:]

print(f"Half 1: {len(half1)} cặp | Half 2: {len(half2)} cặp")

# 4. Tính predictions (cosine similarity) cho từng nửa
def get_predictions(df):
    cosines = []
    for _, row in df.iterrows():
        emb1 = model.encode(row['Sentence1'])
        emb2 = model.encode(row['Sentence2'])
        cos = util.cos_sim(emb1, emb2)[0][0].item()
        cosines.append(cos)
    return np.array(cosines)

pred_half1 = get_predictions(half1)
pred_half2 = get_predictions(half2)

# 5. Tính Spearman correlation giữa 2 nửa
corr, p_value = spearmanr(pred_half1, pred_half2)
print(f"Raw Split-Half Correlation (Spearman): {corr:.4f} (p-value: {p_value:.4f})")

# 6. Áp dụng Spearman-Brown correction để ước lượng reliability toàn bộ
if corr != -1:
    sb_reliability = (2 * corr) / (1 + corr)
    print(f"Split-Half Reliability (Spearman-Brown corrected): {sb_reliability:.4f}")
else:
    print("Correlation âm → reliability thấp hoặc có vấn đề data.")

# 7. Lưu kết quả vào file (tùy chọn)
summary = {
    'Thời gian': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    'Model': model_path,
    'Số cặp test': len(test_df),
    'Raw Correlation': f"{corr:.4f}",
    'Spearman-Brown Reliability': f"{sb_reliability:.4f}",
    'Ghi chú': 'Đánh giá độ tin cậy nội bộ của predictions trên tập test'
}

summary_df = pd.DataFrame([summary])
summary_df.to_csv("split_half_reliability_summary.csv", index=False, encoding='utf-8-sig')
print("\nKết quả đã lưu vào: split_half_reliability_summary.csv")
```
